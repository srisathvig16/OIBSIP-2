```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data1= pd.read_csv("/content/Unemployment in India.csv")
data1.head()
```

```python
data2= pd.read_csv("/content/Unemployment_Rate_upto_11_2020.csv")
data2.head()
```

```python
data1.describe
```

```python
data2.describe
```

```python
plt.figure(figsize=(15, 10))
for i,z in enumerate([' Estimated Unemployment Rate (%)',' Estimated Employed', ' Estimated Labour Participation Rate (%)']):
    plt.subplot(2,2,i+1)
    sns.histplot(x=z, hue="Area", data=data1, kde=True, palette="Set2")
    plt.ylabel("Count")
plt.title("Histogram of Estimated Employment Rate by Area")
plt.show()
```

```python
plt.figure(figsize = (15,6))
sns.boxplot(x='Region', y=' Estimated Unemployment Rate (%)', data=data2,)
plt.title('Box Plot of Estimated Unemployment Rate by Region')
plt.xticks(rotation=45)
plt.show()
```
